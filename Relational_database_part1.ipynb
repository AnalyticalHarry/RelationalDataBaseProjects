#### 1. Database is used to efficiently store and manage data. Database is a well organised electronic filling cabinet that has the ability to store, order and manipulate data while also providing easy access to the user. Database are shared, integrated computer structures that store a collection of end-user data. 

#### 2. DBMS are powerful software, which is known as database management system (DBMS), Which helps manage the contents of the cabinet. A database management system (DBMS) is a collection of programs, which works to manage the database structure and control access to the data stored in the database. 

The DBMS servers as an intermediary between the user and the database.
it receives all application requests and translates the into complex operations required to fulfil those requests.

Much of the database’s internal complexity is hidden from the application programs and end-users by the DBMS. There are some very important advantages to having a DBMS between the end-users application and the database. Firstly, the DBMS allows the data in the database to be shared among multiple applications or users. Secondly, the DBMS integrates many different users’ views of the data into a single data repository.

● Better data sharing: By managing the database and controlling access to data within it, the DBMS makes it possible for end-users to have more efficient access to data that is better managed.

● Improved data integration: By facilitating a variety of end-users to access data in a well-managed manner, the DBMS helps provide a clearer and more integrated view of the organisation’s operations to the end-users.

● Minimised data inconsistency: Data inconsistency occurs when different versions of the same data appear in different places. A properly designed database greatly reduces the probability of data inconsistency as data is drawn from a variety of sources or end-users.

● Improved data access: A query is a specific request for data manipulation (e.g. to read or update the data) sent to the DBMS. The DBMS makes it possible to produce quick answers to spur-of-the-moment queries.

● Improved decision making: Better quality information (on which decisions are made) is generated due to better-managed data and improved data access.

● Increased end-user productivity: The availability of data and the ability to transform data into usable information encourages end-users to make quicker and more informed decisions.

#### 5. A SQL (Structured Query Language) database typically consists of several different types of files, including:

Data files: These files store the actual data of the database, including the rows and columns of each table. They are usually stored in a proprietary format that is specific to the database management system (DBMS) being used.

Log files: These files keep track of all the changes made to the database, including insertions, updates, and deletions. Log files are used to recover the database in case of a crash or other failure.

Index files: These files are used to improve the performance of the database by providing faster access to specific rows of data. They are typically created on one or more columns of a table, and consist of a sorted list of the values in those columns along with a pointer to the corresponding data row.

Control files: These files store information about the physical structure of the database, including the location of data files, log files, and index files. They are used by the DBMS to find and access the various files that make up the database.

Backup files: These files are copies of the database files that are created for the purpose of recovery in case of data loss.

Configuration files: These files contains settings and preferences for the database management system.

Data redundancies refer to the duplication or repetition of data in a database or system. This can lead to inconsistencies and inaccuracies in the data, as well as wasted storage space. It can also make it more difficult to update and maintain the data, as changes may need to be made in multiple places. To avoid data redundancy, database designers often use techniques such as normalization, which involves breaking down data into smaller, more manageable pieces and reducing duplicate data.

#### 6. Data redundancy can cause several issues, including:

Inconsistency: When the same data is stored in multiple locations, it can become difficult to ensure that the data remains consistent across all copies. This can lead to errors and inaccuracies in the data.

Data integrity: When the same data is stored in multiple locations, it can become difficult to maintain the integrity of the data, as changes to one copy may not be reflected in all other copies.

Data duplication: When the same data is stored in multiple locations, it can lead to data duplication, which can result in increased storage costs and decreased efficiency.

Data security: When the same data is stored in multiple locations, it can be difficult to secure all copies of the data, increasing the risk of data breaches.

Performance: Data redundancy can also lead to poor performance, as the database management system must search multiple locations to find the data it needs.

Maintenance: Data redundancy can increase the cost of maintaining the database, as changes must be made to multiple copies of the same data.

To avoid these issues, it is recommended to use a database management system that supports the normalization of data and the use of referential integrity constraints.

### What is anomaly?
Unfortunately, there is no clear definition of the anomaly that’s why I’ve chosen the following one — anomalies are patterns in the data that do not conform to a well-defined notion of normal behaviour [1]. We can illustrate anomalies in a simple two-dimensional space (Figure 1).

N₁ and N₂ are regions with normal data because the majority of observations are in these areas. Points that are far away from normal areas, like points O₁, O₂ and region O₃ are anomalies. Anomalies in the data can occur for different reasons. Malicious activity, credit-card fraud, intrusions, system breakdowns, etc. These anomalies are attractive to the data analyst. Therefore, anomaly detection is an important process and is considered as an advantage in various decision-making systems.

#### Anomalies can be classified into the following three categories:

1. Point Anomalies. If one object can be observed against other objects as anomaly, it is a point anomaly. This is the simplest anomaly category and a lot of researches include them. Taking into consideration example represented in Figure 1 points O₁ and O₂ are point anomalies

2. Contextual Anomalies. If object is anomalous in some defined context. Only in this case it is contextual anomaly (also known as conditional anomaly [2]). In Figure 2 can be seen periodic context. In this case point O₁ is anomaly, because it differs from periodic context

3. Collective Anomalies. If some linked objects can be observed against other objects as anomaly. Individual object can’t be anomalous in this case, only collection of objects.

We need to understand, that these anomalies can be connected. Point anomaly could become contextual if we apply context to it. Or point anomalies could become collective ones, if we join multiple point anomalies together.

On the abstract level detection of the anomalies seems like a simple task. But this task can be very challenging. Here are some challenges bellow:

Defining normal regions is very difficult. In many cases boundaries between anomalies and normal data are not precise. In this case, normal observations could be considered as anomalies and vice-versa

If action is malicious, like fraud, it is considered as anomaly. Very often attackers try to adapt their actions to the normal behaviour. And again, task to identify anomalies in this case is not so simple

What is considered normal today can be not normal in the future. Most of the business systems change in time under the influence of the various factors

Approaches for anomaly detection in one field more often cannot be used in the other one. They will be ineffective in most cases

Training and validation data availability for model training is a big problem


#### Update Anomalies:
    
An update anomaly is a type of database anomaly that occurs when a database update operation results in inconsistent or incorrect data. This can happen when a database update is made to multiple rows or tables at the same time, and not all of the updates are performed correctly or in the correct order. Update anomalies can lead to data integrity issues, and can be difficult to detect and correct. It is important to design and implement proper database constraints, triggers and stored procedures to avoid these anomalies as much as possible.

#### Insertion Anomalies:

There are problems that arise when trying to insert new data into a database. These anomalies can occur when the database design is not properly normalized, resulting in issues such as data redundancy, data inconsistencies, and difficulties updating data. They can also arise when trying to insert new data that does not fit within the constraints of the database schema.

#### Deletion anomalies 

There are problems that arise when deleting data from a database. These anomalies can occur when the database design is not properly normalized, resulting in the loss of important information or data inconsistencies. For example, when deleting a record that is referenced by other records, it can lead to loss of information or unexpected consequences in other parts of the database. This can also occur when deleting a record that contains information that is not stored elsewhere in the database, leading to loss of important data.
